COLUMN           OLD WAY              NEW WAY
────────────────────────────────────────────────────────────

monthly_rent   global median    →   0 if own/family house
                                    renter median if rented

credit_score   global median    →   300 (no history = risky)
               (~680 = good!)       flag column created

bank_balance   global median    →   salary bracket median
                                    (similar earners)
                                    flag column created

emergency_fund global median    →   0 (null = no fund)
                                    flag column created

education      mode (Graduate)  →   'Unknown' category
                                    flag column created

🚩 NEW FLAG COLUMNS ADDED

credit_score_missing    → 1 if credit score was null

bank_balance_missing    → 1 if bank balance was null

emergency_fund_missing  → 1 if emergency fund was null

education_missing       → 1 if education was null

WHY FLAGS MATTER:
Model sees BOTH the filled value AND the flag
It learns: "when flag=1, be more careful"
This is what separates production ML
from notebook ML ✅
